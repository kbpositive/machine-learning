```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs = np.eye(5)
targets = np.random.rand(len(inputs),8)
weights = np.random.rand(len(inputs), len(targets[0]))
```

```python
weights = weights - np.tensordot(inputs,((inputs @ weights)-targets),1)*0.5
fig, axs = plt.subplots(1,len(inputs), figsize=(3*len(inputs), 3))
for index,item in enumerate(axs):
    item.plot(np.array([(inputs @ weights)[index],targets[index]]).T)
```
